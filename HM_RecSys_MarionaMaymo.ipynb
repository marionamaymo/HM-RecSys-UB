La matrix solo nos incluirá los custmers que han comprado al menos uno de los productos que se han comprado al menos 100 veces

Renombramos las columnas y nos quedamos solo con customer_id_x y article_id	

```python
df5 = df4.drop(columns= ['customer_id_y','Count_y'])
df5 = df5.rename(columns={'customer_id_x':'customer_id', 'Count_x':'count'})
df5
```

Ahora creamos el matrix

```python
matrix = df5.pivot_table(
  index='customer_id',
  columns='article_id',
  values='count',
  aggfunc='sum'
)

matrix
```

A continuación lo pasamos a binario

```python
matrix2 = matrix.applymap(lambda x:1 if x>0 else 0)
matrix2
```

Creación modelo colaborativo

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
matrix3_user_user = pd.DataFrame(cosine_similarity(matrix2))
matrix3_user_user 
```

Ahora creamos un df con el resultado de esta matrix y lo guardamos poniéndole un índice

```python
matrix3_user_user.columns= matrix2.index
matrix3_user_user 
```

```python
matrix3_user_user['customer_id'] = matrix2.index
matrix3_user_user = matrix3_user_user.set_index('customer_id')
matrix3_user_user
```
